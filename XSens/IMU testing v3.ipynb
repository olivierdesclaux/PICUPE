```python
import numpy as np
import matplotlib.pyplot as plt
import xsensdeviceapi as xda
import sys
from threading import Lock
import re
import time
import threading as th
import os
import pickle
import datetime
```

```python
f_path = os.path.dirname(os.path.realpath("__file__")) + "\\MTw Pickle"

files = os.listdir(f_path)
print(files)
dataPacket = []

with (open(os.path.join(f_path,files[0]), 'rb')) as openfile:
    while True:
        try:
            dataPacket.append(pickle.load(openfile))
        except EOFError:
            break
```

```python
acc  = [item[1] for item in dataPacket]

acc_x = np.zeros([len(acc),1])
acc_y = np.zeros([len(acc),1])
acc_z = np.zeros([len(acc),1])

for i in range(len(acc)):
    acc_x[i] = acc[i][0]
    acc_y[i] = acc[i][1]
    acc_z[i] = acc[i][2]


fig = plt.figure(figsize=(10, 10))
plt.plot(acc_x, label = "Acceleration X")
plt.plot(acc_y, label = "Acceleration Y")
plt.plot(acc_z, label = "Acceleration Z")
plt.legend()
```

```python
tim= [item[3] for item in dataPacket]
t = tim[4][11:-1].split('.')[0]
x = time.strptime(t,'%H:%M:%S')
s = datetime.timedelta(hours=x.tm_hour,minutes=x.tm_min,seconds=x.tm_sec).total_seconds()
n = int(tim[4][11:-1].split('.')[1])/100
t1 = tim[-1][11:-1].split('.')[0]
x1 = time.strptime(t1,'%H:%M:%S')
s1 = datetime.timedelta(hours=x1.tm_hour,minutes=x1.tm_min,seconds=x1.tm_sec).total_seconds()
n1 = int(tim[-1][11:-1].split('.')[1])/100

print('Time of recording: ',round((s1 +n1)-(s+n),2), 's')
```
